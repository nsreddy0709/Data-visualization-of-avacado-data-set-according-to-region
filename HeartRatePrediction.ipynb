```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.model_selection import train_test_split
dataset = pd.read_csv("/content/drive/My Drive/heart.csv")
dataset.head()
```

```
f1 = np.array(dataset["age"]).reshape(-1,1)
f2 = np.array(dataset["sex"]).reshape(-1,1)
f3 = np.array(dataset["cp"]).reshape(-1,1)
f4 = np.array(dataset["chol"]).reshape(-1,1)
predictor = np.concatenate((f1,f2,f3,f4),axis=1)
predictor
```

```
target = dataset["thalach"]
target
```

```
poly = PolynomialFeatures(13).fit_transform(predictor)
reg = LinearRegression()
reg.fit(poly,target)
```

```
plt.plot(predictor)
plt.plot(reg.predict(poly))
```

```
plt.plot(target)
plt.plot(reg.predict(poly))
```

```
heart_rate_pred = np.array(['63.0','1.0','3.0','233.0'],dtype=np.float32)
poly1 = PolynomialFeatures(13).fit_transform(heart_rate_pred.reshape(1,-1))
reg.predict(poly1)
```
